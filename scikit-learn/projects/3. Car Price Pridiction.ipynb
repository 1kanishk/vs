### DATA CLEANING

```python
car = car[car['year'].str.isnumeric()] #only select rows which have numeric values
```

```python
car['year'] = car['year'].astype(int) #year is in object changing it to integer
```

```python
car = car[car['Price']!='Ask For Price'] #price column has ask for price removing it
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int) #Price has commas in its prices and is in object
```

```python
car['kms_driven'] = car['kms_driven'].str.split().str.get(0).str.replace(',','') #kms_driven has object values with kms at last.
```

```python
car = car[car['kms_driven'].str.isnumeric()] #It has nan values and two rows have 'Petrol' in them
```

```python
car['kms_driven'] = car['kms_driven'].astype(int) #obj to int
```

```python
car.info()
```

```python
car = car[~car['fuel_type'].isna()] #fuel_type has nan values excluding nan values using tilda(~) 
```

```python
#name and company had spammed data...but with the previous cleaning, those rows got removed.
```

```python
car['name'] = car['name'].str.split().str.slice(start=0,stop=3).str.join(' ') #Changing car names. Keeping only the first three words
```

```python
car.reset_index(drop=True)
car
```

```python
car.describe()
```

```python
car[car['Price']<6e6]
car.reset_index(drop=True,inplace=True)
```

```python
car.to_csv('cleaned car.csv')
```

### MODEL

```python
x = car.drop(columns='Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2)
```

```python
from sklearn. linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
cat_columns = ['name', 'company', 'fuel_type']
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), cat_columns),
     remainder = 'passthrough'
)
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
